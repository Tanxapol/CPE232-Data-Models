# Vict Age

The provided Python code is using the pandas library to create a cross-tabulation of the 'Vict Age' column in a DataFrame named `data`.

The `pd.crosstab()` function is a tool in pandas that is used to create a cross-tabulation, which is a type of table in a matrix format that displays the frequency distribution of the variables. It is essentially a summary table that allows you to see the relationship between two or more categorical variables.

In this case, `pd.crosstab(data['Vict Age'], columns = 'N')` is creating a cross-tabulation of the 'Vict Age' column. The `columns = 'N'` argument is specifying that we want to count the frequency of each unique value in the 'Vict Age' column. The result is a DataFrame where the index is the unique values in the 'Vict Age' column, and there is one column named 'N' that contains the count of each unique value.

So, this line of code is creating and returning a frequency distribution of the 'Vict Age' column in the `data` DataFrame. This is a common operation in data analysis when you want to investigate the distribution of categorical data.

```python
pd.crosstab(data['Vict Age'], columns = 'N')
```

# Crm Cd Desc

This code uses the pd.crosstab function from the pandas library to create a contingency table (cross-tabulation). It uses data from the 'Crm Cd Desc' column in your DataFrame. The resulting table is stored in the variable Crm_Desc. After running this code, you can view the generated contingency table by typing Crm_Desc.

```python
Crm_Desc = pd.crosstab(data['Crm Cd Desc'], columns = 'N')
Crm_Desc
# Crm_Desc.to_csv('table.csv', index=True)
```

It groups the DataFrame 'data' by the 'Crm Cd Desc' column.
For each group, it extracts the unique values from the 'Crm Cd' column using the unique() method.
It applies a lambda function to convert the unique values into a string separated by commas using ', '.join(map(str, x)).
The result is stored in the variable Crm.
Finally, if you uncomment the line # Crm.to_csv('Crm.csv', index=True), it will save the resulting DataFrame to a CSV file named 'Crm.csv' with the index included.

```python
Crm = data.groupby('Crm Cd Desc')['Crm Cd'].unique().apply(lambda x: ', '.join(map(str, x))).reset_index()
Crm
# Crm.to_csv('Crm.csv', index=True)
```

### **Age Group Analysis**

*   กำหนด bin และ label สำหรับกลุ่มอายุต่าง ๆ
*   สร้างคอลัมน์ใหม่ 'Age Group' ตามคอลัมน์ 'Vict Age'
*   ใช้ `pd.crosstab` เพื่อสร้าง cross-tabulation ของคอลัมน์ 'Age Group'
*   แสดง result จากการแยกตาม 'Age Group'



```python
data1 = data[data['Crm Cd Desc'] == 'ROBBERY']

bins = [1, 11, 25, 41, 57, 76, 130]
labels = ['Gen Alpha', 'Gen Z', 'Gen Y', 'Gen X', 'Baby Boomer', 'Silent Gen']

data1['Age Group'] = pd.cut(data1['Vict Age'], bins=bins, labels=labels)

result = pd.crosstab(data1['Age Group'], columns='N')
result
```

```python
result.plot(kind='bar', legend=False)
plt.xlabel('Age Group')
plt.ylabel('Frequency')
plt.title('Frequency of Age Groups')
plt.show()
```

### **Premises Description Analysis**

*   กรองข้อมูล data1 สำหรับ Gen Y
*   สร้าง cross-tabulation ของคอลัมน์ 'Premis Desc' สำหรับ Gen Y และทำการแสดงผล

Find the maximum count and corresponding value
*   หา index max หรือ index ที่มีการเกิดขึ้นสูงสุด จาก premresult_gen_y
*   หาค่า max หรือ ค่าที่มีการเกิดขึ้นสูงสุด จาก premresult_gen_y



```python
gen_y_data = data1[data1['Age Group'] == 'Gen Y']

premresult_gen_y = pd.crosstab(gen_y_data['Premis Desc'], columns='N')

print(premresult_gen_y)
```

```python
# Find the maximum count and corresponding value
max_count = premresult_gen_y['N'].max()
max_value = premresult_gen_y['N'].idxmax()

# Print the result
print(f"Maximum Count: {max_count}")
print(f"Corresponding Value: {max_value}")
```

```python
premresult_gen_y.plot(kind='bar', legend=False)
plt.xlabel('Premis Desc')
plt.ylabel('N')
plt.title('Places where Gen Y are victims of robbery')
plt.show()
```
